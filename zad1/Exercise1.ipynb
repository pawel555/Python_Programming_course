```
def caught_speeding(speed, is_birthday):
  additional_limit = 0
  if is_birthday is True:
    additional_limit = 5
  if speed <= 60 + additional_limit:
    return 'No ticket'
  elif 60 + additional_limit <= speed <= 80 + additional_limit:
    return 'Small Ticket'
  elif speed >= 80 + additional_limit:
    return 'Big Ticket'
  
  

print(caught_speeding(59,False))
print(caught_speeding(61,True))
print(caught_speeding(61,False))
print(caught_speeding(64,True))
print(caught_speeding(70,False))
print(caught_speeding(81,True))
print(caught_speeding(81,False))
```
